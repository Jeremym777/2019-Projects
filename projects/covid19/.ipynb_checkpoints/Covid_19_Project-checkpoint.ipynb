### 1. Data Preparation


```python
US_confirm=us_confirmed_cases.iloc[:, np.r_[6, 11:len(us_confirmed_cases.columns)]].groupby("Province_State",as_index=True).sum()
US_confirm['Confirmed']=US_confirm[['4/18/20']]
US_confirm=US_confirm[['Confirmed']]
US_confirm=US_confirm.reset_index()
US_confirm.head()
```

-Above is the total number of confirmed cases for each state/territory.


```python
US_deaths=us_confirmed_deaths.iloc[:, np.r_[6, 11:len(us_confirmed_deaths.columns)]].groupby("Province_State",as_index=True).sum()
US_deaths['Deaths']=US_deaths[['4/18/20']]
US_deaths['New Pop']=US_deaths['Population']-US_deaths['Deaths']
US_deaths=US_deaths[['Population','Deaths','New Pop']]
US_deaths=US_deaths.reset_index()
US_deaths.head()
```

- Above is the total number of deaths for each state and territory with the original population before deaths.

```python
US_recovery=global_states[global_states['Country_Region']=='US']
US_recovery=US_recovery[['Province_State','Country_Region','Active','Recovered']]
US_recovery=US_recovery.replace(np.nan,0)
US_recovery=US_recovery[:-1]
US_recovery.head()
```

- Above is all the recovered and active cases in the US


```python
US_tested=global_states[global_states['Country_Region']=='US']
US_tested=US_tested[['Province_State','Country_Region','People_Tested','Incident_Rate','Hospitalization_Rate','Mortality_Rate','People_Hospitalized']]
US_tested=US_tested.replace(np.nan,0)[:-1]
US_tested.head()
```

Above are rates, people tested and hospitalized.

```python
US_Data=pd.merge(US_confirm,US_deaths,'inner')
US_Data=pd.merge(US_Data,US_recovery,'inner')
US_Data=pd.merge(US_Data,US_tested,'inner')
US_Data=US_Data[['Province_State','Country_Region','Population','Deaths','New Pop','Active','Recovered',
                 'People_Tested','Incident_Rate','Hospitalization_Rate','Mortality_Rate','People_Hospitalized']]
US_Data.head()
```

- All Data for the US/Territories placed into one data set.

```python
global_confirmed['Total Confirmed']=global_confirmed[['4/27/20']]
confirm=global_confirmed[['Country/Region','Lat','Long','Total Confirmed']]

death=global_deaths.iloc[:, np.r_[1, 4:len(global_deaths.columns)]]
death['Total Deaths']=global_deaths[['4/27/20']]
death=death[['Country/Region','Total Deaths']]

global_recoveries['Total Recovery']=global_recoveries[['4/27/20']]
recoveries=global_recoveries[['Country/Region','Lat','Long','Total Recovery']]
```

```python
test=np.array(confirm['Country/Region'])
test1=np.array(confirm['Lat'])
test2=np.array(confirm['Long'])
test3=np.array(confirm['Total Confirmed'])
test4=np.array(death['Total Deaths'])
global_totals=pd.DataFrame({'Country/Region':test,
                            'Lat':test1,
                            'Long':test2,
                            'Total Confirmed': test3,
                            'Total Deaths': test4})
global_totals=pd.merge(global_totals,recoveries,on=['Country/Region','Lat','Long'],how='left')
global_totals=global_totals.replace(np.nan,0)
global_totals
```

#2. Visualization

```python
top_20_confirmed=US_confirm.reset_index().sort_values('Confirmed',ascending=False).head(20)
fig = plt.figure(figsize=(10,8))
ax = fig.add_axes([0,0,1,1])
plt.title("Top 20 Confirmed Cases in the U.S. (4/18/20)",fontsize=20)
plt.xlabel('Number of Cases',fontsize=20)
plt.ylabel('State',fontsize=20)
ax.barh(top_20_confirmed["Province_State"], top_20_confirmed["Confirmed"],color='blue')
plt.yticks(fontsize=8)
ax.invert_yaxis()
plt.show()
```

```python
Top_20_deaths=US_deaths.sort_values('Deaths',ascending=False,inplace=False).head(20)
fig = plt.figure(figsize=(10,8))
ax = fig.add_axes([0,0,1,1])
plt.title("Top 20 Deaths in the U.S. (4/18/20)",fontsize=20)
plt.xlabel('Number of Deaths',fontsize=20)
plt.ylabel('State',fontsize=20)
ax.barh(Top_20_deaths["Province_State"], Top_20_deaths["Deaths"],color='red')
plt.yticks(fontsize=8)
ax.invert_yaxis()
plt.show()
```

```python
top_20_recoveries=US_recovery.sort_values('Recovered',ascending=False).head(20)
fig = plt.figure(figsize=(10,8))
ax = fig.add_axes([0,0,1,1])
plt.title("Top 20 Recovered in the U.S. (4/18/20)",fontsize=20)
plt.xlabel('Recovered',fontsize=20)
plt.ylabel('State',fontsize=20)
ax.barh(top_20_recoveries["Province_State"], top_20_recoveries["Recovered"],color='yellow')
plt.yticks(fontsize=8)
ax.invert_yaxis()
plt.show()
```

```python
# imports Basemap
from mpl_toolkits.basemap import Basemap
%matplotlib inline
```

```python
colors=pd.factorize(global_totals['Country/Region'])[0]
colors
def plt_confirm(df):
    # dim of fig
    plt.figure(figsize=(20, 18), dpi=96)

    # Creates the backgroud map
    m=Basemap(llcrnrlon=-180, llcrnrlat=-65,urcrnrlon=180,urcrnrlat=80)
    m.drawmapboundary(fill_color='#A6CAE0', linewidth=0)
    m.fillcontinents(color='green', alpha=0.3)
    m.drawcoastlines(linewidth=0.1, color="white")
    
    total_cases = np.sum(df['Total Confirmed'])

    # Adds pt based on location
    m.scatter(df['Long'], 
              df['Lat'], 
              s = df['Total Confirmed'] * .025, # Controls circle size, can change to get bigger or smaller circles
              alpha=0.4, 
              c=colors, 
              cmap="Set1")
    plt.title('World Confirmed Covid-19 Cases As of 4/27/20: ' + str(int(total_cases)) + '\n(circles not to scale)', fontsize=30)
    plt.show()
    return
plt_confirm(global_totals)
```

```python
def plt_deaths(df):
    # dim of fig
    plt.figure(figsize=(20, 18), dpi=96)

    # Creates the backgroud map
    m=Basemap(llcrnrlon=-180, llcrnrlat=-65,urcrnrlon=180,urcrnrlat=80)
    m.drawmapboundary(fill_color='#A6CAE0', linewidth=0)
    m.fillcontinents(color='green', alpha=0.3)
    m.drawcoastlines(linewidth=0.1, color="white")
    
    total_deaths = np.sum(df['Total Deaths'])

    # Adds pt based on location
    m.scatter(df['Long'], 
              df['Lat'], 
              s = df['Total Deaths'] * .3, # Controls circle size, can change to get bigger or smaller circles
              alpha=0.4, 
              c=colors, 
              cmap="Set1")
    plt.title('World Deaths from Covid-19 As of 4/27/20: ' + str(int(total_deaths)) + '\n(circles not to scale)', fontsize=30)
    plt.show()
    return
plt_deaths(global_totals)
```

```python
def plt_recoveries(df):
    # dim of fig
    plt.figure(figsize=(20, 18), dpi=96)

    # Creates the backgroud map
    m=Basemap(llcrnrlon=-180, llcrnrlat=-65,urcrnrlon=180,urcrnrlat=80)
    m.drawmapboundary(fill_color='#A6CAE0', linewidth=0)
    m.fillcontinents(color='green', alpha=0.3)
    m.drawcoastlines(linewidth=0.1, color="white")
    
    total_cases = np.sum(df['Total Recovery'])

    # Adds pt based on location
    m.scatter(df['Long'], 
              df['Lat'], 
              s = df['Total Recovery'] * .2, # Controls circle size, can change to get bigger or smaller circles
              alpha=0.4, 
              c=colors, 
              cmap="Set1")
    plt.title('World Recoveries From Covid-19 As of 4/27/20: ' + str(int(total_cases)) + '\n(circles not to scale)', fontsize=30)
    plt.show()
    return
plt_recoveries(global_totals)
```
